```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
f_x = np.array([0.1, 0.2, 0.5, 0.8, 1, 2, 5, 8, 15, 25, 50, 75, 100])
U_y = np.array([2.98, 2.93, 2.93, 2.94, 2.93, 2.94, 7.82, 9.12, 8.84, 6.34, 3.03, 2.95, 2.94])
```

```python
plt.figure(figsize=(9.0, 9.0), dpi=600)
plt.grid()
plt.plot(f_x, U_y)
plt.title(" ")
plt.xlabel("f(V1), ")
plt.ylabel(", ")
plt.xlim(0, 100)
plt.ylim(0, 10)
plt.show()
```

```python
L = 6.25e-3
C = 100e-9
R = 1.7e+3
```

```python
fp = 1 / ( 2 * np.pi * np.sqrt(L * C) )
```

```python
p = lambda f: f / fp
```

```python
K = lambda f: ((p(f) * R * C) / (1 + p(f) * R * C + p(f)**3 * L * C))
```

```python
freq = np.arange(100, (10e+3 + 1), 100)
U_an = K(freq)
print(U_an)
```

```python
plt.figure(figsize=(9.0, 9.0), dpi=600)
plt.grid()
plt.plot(freq, U_an)
plt.title(" ")
plt.xlabel("f(V1), ")
plt.ylabel(", ")
#plt.xlim(0, 100)
#plt.ylim(0, 10)
plt.show()
```
